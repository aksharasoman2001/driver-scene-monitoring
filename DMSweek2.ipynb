```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np

raw_path = '/content/drive/MyDrive/DMS_Project/data/raw'
open_eye_path = os.path.join(raw_path, 'open_eye')
closed_eye_path = os.path.join(raw_path, 'closed_eye')

# Get image lists
open_images = os.listdir(open_eye_path)
closed_images = os.listdir(closed_eye_path)

# Plot sample images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('Dataset Samples — CEW Dataset', fontsize=16)

# Show 5 open eye images
for i in range(5):
    img_path = os.path.join(open_eye_path, open_images[i])
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[0, i].imshow(img)
    axes[0, i].set_title('Open Eye')
    axes[0, i].axis('off')

# Show 5 closed eye images
for i in range(5):
    img_path = os.path.join(closed_eye_path, closed_images[i])
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[1, i].imshow(img)
    axes[1, i].set_title('Closed Eye')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()

# Print summary
print(f"\n📊 Dataset Summary:")
print(f"Open eye images:   {len(open_images)}")
print(f"Closed eye images: {len(closed_images)}")
print(f"Total images:      {len(open_images) + len(closed_images)}")
print(f"Class balance:     {len(open_images)/len(closed_images):.2f}")
print(f"Image size:        136 x 136 x 3")
```

```python
import os

open_eye_path = '/content/drive/MyDrive/DMS_Project/data/raw/open_eye'
closed_eye_path = '/content/drive/MyDrive/DMS_Project/data/raw/closed_eye'

# Build file paths and labels
all_paths = []
all_labels = []

# Add open eye images → label 1
for img_name in os.listdir(open_eye_path):
    all_paths.append(os.path.join(open_eye_path, img_name))
    all_labels.append(1)

# Add closed eye images → label 0
for img_name in os.listdir(closed_eye_path):
    all_paths.append(os.path.join(closed_eye_path, img_name))
    all_labels.append(0)

print(f"Total images:  {len(all_paths)}")
print(f"Total labels:  {len(all_labels)}")
print(f"Sample path:   {all_paths[0]}")
print(f"Sample label:  {all_labels[0]}")
```

```python
from sklearn.model_selection import train_test_split

# First split — separate test set (15%)
train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
    all_paths, all_labels,
    test_size=0.15,
    random_state=42,
    stratify=all_labels
)

# Second split — separate val from train (15% of remaining)
train_paths, val_paths, train_labels, val_labels = train_test_split(
    train_val_paths, train_val_labels,
    test_size=0.176,
    random_state=42,
    stratify=train_val_labels
)

print(f"Train: {len(train_paths)}")
print(f"Val:   {len(val_paths)}")
print(f"Test:  {len(test_paths)}")
```

```python
from torchvision import transforms

# For val and test — no augmentation
base_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

print("Base transform ready")
```

```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
import numpy as np

train_transform = A.Compose([
    A.Resize(64, 64),
    A.RandomBrightnessContrast(p=0.5),
    A.MotionBlur(blur_limit=5, p=0.3),
    A.HorizontalFlip(p=0.5),
    A.GaussNoise(p=0.3),
    A.CoarseDropout(num_holes_range=(1, 1),
                    hole_height_range=(20, 20),
                    hole_width_range=(20, 20), p=0.3),
    A.ToGray(p=0.1),
    A.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

print("Train transform ready")
```

```python
from torch.utils.data import DataLoader

train_dataset = EyeDataset(train_paths, train_labels, transform=train_transform, use_albumentations=True)
val_dataset   = EyeDataset(val_paths,   val_labels,   transform=base_transform,  use_albumentations=False)
test_dataset  = EyeDataset(test_paths,  test_labels,  transform=base_transform,  use_albumentations=False)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=32, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=32, shuffle=False)

print(f"Train batches: {len(train_loader)}")
print(f"Val batches:   {len(val_loader)}")
print(f"Test batches:  {len(test_loader)}")
```

```python
import os
from torch.utils.data import Dataset
from PIL import Image
import numpy as np

class EyeDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None, use_albumentations=False):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform
        self.use_albumentations = use_albumentations

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        if self.transform:
            if self.use_albumentations:
                img = np.array(img)                       # albumentations needs numpy
                img = self.transform(image=img)["image"]  # returns a dict
            else:
                img = self.transform(img)                 # torchvision works on PIL
        return img, self.labels[idx]

print("EyeDataset class ready")
```
